```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Algorithm -
#Orb Feature Matcher

def getOrbCorrespondences(img):
    
    #It can get at most 10k features.
    orb = cv2.ORB_create(nfeatures = 10000)
    kp = orb.detect(img,None)
    kp, des = orb.compute(img, kp)
    
    #Return keypoints and destinations
    return kp, des

def getMatches(des1,des2):
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    matches = bf.match(des1,des2)
    matches = sorted(matches, key = lambda x:x.distance)

    # Draw matches.
    img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:50],None)
    plt.imshow(img3), plt.show()
    return matches

def drawDisparity(kp1,kp2,matches,realDisparity):

    result = 255 * np.ones(shape=[512, 512, 3], dtype=np.uint8)

    for i in range(0,len(matches)):
        p1 = kp2[matches[i].trainIdx].pt
        p2 = kp1[matches[i].queryIdx].pt

        #This is eliminates wrong matches, because we only scan horizontal displacement.
        if p1[1] - p2[1] != 0: 
            continue

        x = p2[0]
        y = p2[1]

        #We get a distance metric from there, it helps to colorize our map!
        colorCode = abs(p1[0] - p2[0])*5
        if colorCode > 200:
            colorCode = 200

        cv2.rectangle(result, (int(x),int(y)),(int(x+10),int(y+10)), color=(colorCode,colorCode,colorCode), thickness=-10)

    #Show the result.
    print("Calculated result:")
    plt.imshow(result)
    plt.show()

    print("Real result from middlebury:")
    plt.imshow(realDisparity)
    plt.show()
```

```python
#TEST RUN 1 : bull
img1 = cv2.imread("saw1.ppm")
img2 = cv2.imread("saw2.ppm")
realDisparity = cv2.imread("DISPsaw.pgm")

kp1, des1 = getOrbCorrespondences(img1)
kp2, des2 = getOrbCorrespondences(img2)

matches = getMatches(des1,des2)

drawDisparity(kp1,kp2,matches,realDisparity)
#IF IT GIVES AN ERROR, IT DOES NOT FIND THE IMAGE TO SCAN...MAKE SURE YOU GIVE RIGHT NAME!
```
